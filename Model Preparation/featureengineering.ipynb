## In this model we will be performing the feature engineering. 

```python
### Libraries we will be using 

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
## Dataset we will be using 

data = pd.read_csv('../Datasets/cleaned.csv')
data.drop(columns=['Unnamed: 0','customerID'], inplace=True)
data.head(5)
```

```python
### Analyzing the data types of each feature

print(data.dtypes)
```

### We have to convert all the columns with str dtype to either integer or float.

```python
strdtypes = data.select_dtypes('str')
for col in strdtypes.columns:
    print(col, '->', strdtypes[col].unique())
```

```python
### Converting each column 
data['gender'] = data['gender'].map({'Male':1,'Female':0})

## Columns with yes or no values 
yncolumns = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

for each in yncolumns:
    data[each] = data[each].map({'Yes':1,'No':0})

## We have plotted all the yes and no values to 1 and 0 

## Columns with yes,no,nointernetservice

ynncolumns = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies']

for each in ynncolumns:
    data[each] = data[each].apply(lambda x: 1 if x == 'Yes' else 0)


### Payment method 

mapping_payment = {'Electronic check':0, 'Mailed check':1,'Bank transfer (automatic)':2,   'Credit card (automatic)':3  }

data['PaymentMethod'] = data['PaymentMethod'].map(mapping_payment)

## Converting the Contract 

mapping_contract = {'Month-to-month':0, 'One year':1, 'Two year':2 }
data['Contract'] = data['Contract'].map(mapping_contract)

### Converting the Internet Service

mapping_internet = {'DSL':0, 'Fiber optic':1, 'No':2}
data['InternetService'] = data['InternetService'].map(mapping_internet)
```

```python
### Checking the dtypes of all the columns
print(data.dtypes)

```

### We have successfully converted all the string data types to integers
---

```python
### Lets check if the newly created binary dataset has any Null values 

print(data.isna().sum())
```

### There are no null values in the binary dataset, which is a poisitve sign for model deployment. 

---
### Lets check the correlation between the columns

```python
correlated_matrix = data.corr()
correlated_matrix.head()
```

```python
### Upper triangular matrix to reduce the duplicate values 
upper = correlated_matrix.where(np.triu(np.ones(correlated_matrix.shape), k =1 ).astype(bool))
```

### Now lets check the highly correlated columns


```python
high_correlated_features = [column for column in upper.columns if any(upper[column] > 0.95)]
print(high_correlated_features)
```

### We figured out that no columns are highly correlated, which concludes that every column has its own significance.

---
### Variance in the data 

```python
### Checking the variance in the data 
only_numerical = data.select_dtypes(include=['int64', 'float64'])
variance = only_numerical.var()

### Lets check if any of the columns are less varied

print(variance[variance <= 0.05]) ## Variance less than 5%
```

### As we can see that every column are varied, which suggests that there is variation in the data

---

### Checking the skewness of the data

```python
selected_columns = data.drop(columns=['Churn'])
only_numeric_selected = selected_columns.select_dtypes(include=['int64', 'float64'])
skewed_data = only_numeric_selected.skew()
print(skewed_data)
### As seen there is no such skewness in the data 
```
