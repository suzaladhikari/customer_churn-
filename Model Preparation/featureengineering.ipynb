## In this model we will be performing the feature engineering. 

```python
### Libraries we will be using 

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
## Dataset we will be using 

data = pd.read_csv('../Datasets/cleaned.csv')
data.drop(columns=['Unnamed: 0','customerID'], inplace=True)
data.head(5)
```

```python
### Analyzing the data types of each feature

print(data.dtypes)
```

### We have to convert all the columns with str dtype to either integer or float.

```python
strdtypes = data.select_dtypes('str')
for col in strdtypes.columns:
    print(col, '->', strdtypes[col].unique())
```

```python
### Converting each column 
data['gender'] = data['gender'].map({'Male':1,'Female':0})

## Columns with yes or no values 
yncolumns = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

for each in yncolumns:
    data[each] = data[each].map({'Yes':1,'No':0})

## We have plotted all the yes and no values to 1 and 0 

## Columns with yes,no,nointernetservice

ynncolumns = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies']

for each in ynncolumns:
    data[each] = data[each].apply(lambda x: 1 if x == 'Yes' else 0)


### Payment method 

mapping_payment = {'Electronic check':0, 'Mailed check':1,'Bank transfer (automatic)':2,   'Credit card (automatic)':3  }

data['PaymentMethod'] = data['PaymentMethod'].map(mapping_payment)

## Converting the Contract 

mapping_contract = {'Month-to-month':0, 'One year':1, 'Two year':2 }
data['Contract'] = data['Contract'].map(mapping_contract)

### Converting the Internet Service

mapping_internet = {'DSL':0, 'Fiber optic':1, 'No':2}
data['InternetService'] = data['InternetService'].map(mapping_internet)
```

```python
### Checking the dtypes of all the columns
print(data.dtypes)

```
